```python
#import the image
img = cv.imread('D:\AI-datasource\IMG.png')
img1 = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
plt.imshow(img1)
```

```python
#convert the image into gray scale
img_gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
```

```python
# plot the three channels of the image
fig, axs = plt.subplots(nrows = 1,ncols = 3,figsize = (20,20))
for i in range(0,3):
    ax = axs[i]
    ax.imshow(img[:,:,i],cmap = 'gray')
plt.show()    
```

```python
#transform imae into hsv&hsl
img_hsv=cv.cvtColor(img,cv.COLOR_BGR2HSV)
img_hls = cv.cvtColor(img, cv.COLOR_BGR2HLS)
```

```python
#plot the converted images
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(20,20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
# copy image
img_copy= img1.copy()
#draw a rectangle
cv.rectangle(img_copy,pt1=(0,50),pt2=(700,100),
            color = (0,0,0),thickness = -1)
plt.imshow(img_copy)
```

```python
#draw a circle
cv.circle(img_copy,center= (250,275),radius = 60,
         color=(0,0,255),thickness=2)
plt.imshow(img_copy)
```

```python
# add text
cv.putText(img_copy, text= "The Burano Island",
          org = (200,80),
          fontFace= cv.FONT_HERSHEY_DUPLEX,
          fontScale=1,
          color=(255,255,255),
          thickness=2,
          lineType= cv.LINE_AA)
plt.imshow(img_copy)
```
