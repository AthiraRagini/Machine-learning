# Training And Testing Data

```python
b= b.drop(['Posted On'], axis=1)#removes the specified  column
b
```

```python
b= pd.get_dummies(b,drop_first=True)#It converts categorical data into dummy or indicator variables
b
```

```python
X= b.drop('Rent',axis=1)#Arrange Data into Features and Target,
                         #here the features in X and the target in y,for Train Test Split
y=b['Rent']             #here features refers the details about the house and target means the corresponding rent of each.
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=0.3,random_state=2)
```

# MODEL CREATION

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X,train_y) #model creation
reg
```

```python
reg.predict(test_X)
```

```python
reg.score(train_X,train_y)#accuracy
```

```python

reg = LinearRegression()
reg.fit(train_X,train_y)
plt.scatter(reg.predict(train_X),train_y) #plot to  observe relationship between variables ,here the real value of y and the predicted
plt.xlabel('Predicted value of y')      # value of y
plt.ylabel('Real value of y')
plt.show()
```
