```python
import cv2
```

```python
#load cascade
face_cascade=cv2.CascadeClassifier('D:\AI-datasource\haarcascade.xml')
```

```python
# Read the input image
img = cv2.imread('D:\AI-datasource\people.png',cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img)
```

```python
# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
# Detect faces
faces = face_cascade.detectMultiScale(img, 1.1, 4)
```

```python
?face_cascade.detectMultiScale
```

```python
# Detect faces
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1,minNeighbors=4,minSize = (10,10))
```

```python
faces
```

```python
len(faces)
```

```python
# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 0), 2)
```
