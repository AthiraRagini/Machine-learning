# To predict a person salary>100k based on company,job and degree

```python
import pandas as pd
```

```python
data=pd.read_csv('D:\AI-datasource\data13.csv')
data
```

```python
#splitting the dataset into input and output
#input=(job,degree,and company)
#output=salary_more_then_100k
```

```python
input=data.drop(['salary_more_then_100k'],axis=1)
```

```python
input
```

```python
target=data.salary_more_then_100k
```

```python
target #salary_more_then_100k  0==>no, 1==>yes
```

```python
#ML model work very well on numerical data so we have to convert the data of input into numbers 
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
# to  create new columns in input that represents its own numerical form
input['com_n']=le_company.fit_transform(input['company'])
input['job_n']=le_job.fit_transform(input['job'])
input['degree_n']=le_degree.fit_transform(input['degree'])
```

```python
input
```

```python
# drop the next data from inputs
input_n=input.drop(['company','job','degree'],axis=1)
```

```python
target
```

```python
input_n
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

Q: Is salary of google ,computer engineer,bachelors degree>100k?

```python
model.predict([[2,1,0]]) #google=2,computer engineer=1,bacherlors degree=0,output=no
```

Q: Is salary of google ,computer engineer,masters degree>100k?

```python
model.predict([[2,1,1]])
```
